```python
# Import Data
import matplotlib.pyplot as plt 
import numpy as np
import scipy.stats as stat 
import seaborn as sns
sns.set()
```

```python
# Menemtukan rataan 2 kelompok
rataan1 = 50 # dengan pupuk 
rataan2 = 51 # tanpa pupuk

# menentukan standar deviasi (variasi) --sama antar 2 kelompok
sd = 5

# ukuran sampel 2 kelompok
sampel1 = 50
sampel2= 40 
```

```python
# generate bilangan random
tes = np.random.randn(sampel1)
tes
```

```python
plt.plot(tes);
```

```python
sns.distplot(tes);
```

```python
# generate data 
data1 = np.random.randn(sampel1)*sd + rataan1
data2 = np.random.randn(sampel2)*sd + rataan2

# menggabungkan sample size (jumlah data)
kelompok_data = [sampel1,sampel2]

batas = [np.min(np.hstack((data1, data2))), np.max(np.hstack((data1, data2)))]
```

```python
batas
```

```python
# Plot distibrusi kedua kelompok (statistik deskriptif)
plt.figure(figsize=(7,5))

a = sns.distplot(data1, hist=False, label= 'data 1')
a = sns.distplot(data2, hist=False, label= 'data 2')

a.set(xlabel='Data', ylabel= 'Density')
plt.legend();
```

```python
# melakukan uji statistik (statistik inferensial)
fig, ax = plt.subplots(1, 2, figsize=(8, 5), dpi=300)
ax[0].violinplot(data1);
ax[0].plot(1+np.random.randn(sampel1)/10, data1, 'o', color='green');
ax[0].set_ylim(batas)

ax[1].violinplot(data2);
ax[1].plot(1+np.random.randn(sampel2)/10, data2, 'o', color='blue');
ax[1].set_ylim(batas)

# melakukan uji t-test
t, p = stat.ttest_ind(data1, data2)

# Mencetak hasil t-test
sigtxt = ('', 'TIDAK')
plt.title(' dua kelompok {} berbeda secara signifikan! t({})={}, p={}'.format(sigtxt[int(p>.05)], sum(kelompok_data)-2, np.round(t, 2), np.round(p,6)))
plt.show()
```
